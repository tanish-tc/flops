```python
import requests
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from concurrent.futures import ThreadPoolExecutor

payload = {
  "document_id": "doc_10485",
  "ocr_text": "Invoice total is $500. Due Date: May 1 2026. Please remit payment promptly."
}

endpoints = {
    "Baseline (PyTorch)": "http://localhost:8000/predict",
    "Optimized (ONNX)": "http://localhost:8001/predict",
    "Extra Credit (Ray)": "http://localhost:8002/predict"
}

results_data = []
summary_stats = []

def send_req(url):
    s = time.time()
    try:
        res = requests.post(url, json=payload, timeout=5)
        if res.status_code == 200:
            return time.time() - s
    except:
        pass
    return None

for name, url in endpoints.items():
    print(f"Benchmarking {name}...")
    start_total = time.time()
    
    with ThreadPoolExecutor(max_workers=15) as ex:
        latencies = list(ex.map(lambda _: send_req(url), range(300)))
        
    latencies = [l for l in latencies if l is not None] # Filter failures
    
    if len(latencies) == 0:
        print(f"  -> WARNING: All requests to {name} failed. Skipping.\n")
        continue
        
    total_time = time.time() - start_total
    throughput = len(latencies) / total_time
    
    for lat in latencies:
        results_data.append({"Framework": name, "Latency (ms)": lat * 1000})
        
    summary_stats.append({
        "Framework": name, 
        "Throughput (req/sec)": throughput,
        "p95 Latency (ms)": np.percentile(latencies, 95) * 1000
    })

if summary_stats:
    df_raw = pd.DataFrame(results_data)
    df_summary = pd.DataFrame(summary_stats)

    sns.set_theme(style="whitegrid")
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    sns.barplot(data=df_summary, x="Framework", y="Throughput (req/sec)", ax=axes[0], palette="viridis")
    axes[0].set_title("Throughput (Higher is Better)", fontweight='bold')
    axes[0].set_ylabel("Requests per Second")

    sns.boxplot(data=df_raw, x="Framework", y="Latency (ms)", ax=axes[1], palette="viridis")
    axes[1].set_title("Latency (Lower & Tighter is Better)", fontweight='bold')
    axes[1].set_ylabel("Latency (milliseconds)")

    plt.tight_layout()
    plt.show()

    print("\n--- SUMMARY METRICS ---")
    print(df_summary.to_markdown(index=False))
else:
    print("All endpoints failed. No data to graph.")
```
